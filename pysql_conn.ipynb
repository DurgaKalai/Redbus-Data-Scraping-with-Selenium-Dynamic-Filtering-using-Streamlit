```python
#importing required libraries
import pandas as pd
import pymysql
import numpy as np
```

```python
# csv to dataframe
df1=pd.read_csv("Andhra_bus_details.csv")
df2=pd.read_csv("telangana_bus_details.csv")
df3=pd.read_csv("kerala_bus_details.csv")
df4=pd.read_csv("kadamba_bus_details.csv")
df5=pd.read_csv("rajasthan_bus_details.csv")
df6=pd.read_csv("southbengal_bus_details.csv")
df7=pd.read_csv("himachalpradesh_bus_details.csv")
df8=pd.read_csv("uttarpradesh_bus_details.csv")
df9=pd.read_csv("chandigarh_bus_details.csv")
df10=pd.read_csv("punjab_bus_deta.csv")

Final_df=pd.concat([df1,df2,df3,df4,df5,df6,df7,df8,df9,df10],ignore_index=True)
Final_df
```

# Data_cleaning

```python
# details of dataframe structure
Final_df.info()
```

```python
#convert price column to numeric values
Final_df["Price"]=Final_df["Price"].str.replace("INR","")
Final_df["Price"]=Final_df["Price"].astype(float)
Final_df["Price"].fillna(0)
```

```python
#convert Ratings to numeric
Final_df["Star_Rating"] = Final_df["Star_Rating"].astype(str)
Final_df["Star_Rating"]=Final_df["Star_Rating"].str.replace("New","")
Final_df["Star_Rating"]=Final_df["Star_Rating"].str.strip()
Final_df["Star_Rating"]=Final_df["Star_Rating"].str.split().str[0]
Final_df["Star_Rating"] = pd.to_numeric(Final_df["Star_Rating"], errors='coerce')
Final_df["Star_Rating"].fillna(0)
```

```python
# info after the data type change
Final_df.info()
```

```python
# replacing the nan value
Final_df = Final_df.replace({np.nan: None})
```

```python
#adding id column
id_column = pd.Series(range(1, len(Final_df) + 1), name='id')
Final_df = pd.concat([id_column,Final_df ], axis=1)
```

```python
# change dataframe to csv
path=r"C:\Users\DURGA\Desktop\Project Redbus\Final_busdetails_df.csv"
Final_df.to_csv(path,index=False)
```
